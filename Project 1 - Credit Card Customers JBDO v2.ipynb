## Key Questions:
### 1. Comparing gender and age to credit card users that have stayed and left?

```python
# columns needed for this are Gender, Customer_Age, Attrition_Flag
Q1_columns = ["Gender", "Customer_Age", "Attrition_Flag"]
Q1_df = CCCP_data[Q1_columns]
Q1_df
```

```python
# determining the percentage of existing customer vs churned customers
# using the '.count()' function we can determine the number of attritted customers
num_att_customer = CCCP_data[CCCP_data.Attrition_Flag=="Attrited Customer"].Attrition_Flag.count()
# to get the total number of customers the function '.nunique()' was used looking at the CLIENTNUM
tot_num_customer = CCCP_data.CLIENTNUM.nunique()
# number of existing customers can be determined by subtraction
num_exist_customer = tot_num_customer - num_att_customer

# determine the percentages of each customer "base"
att_customer_percent = (num_att_customer / tot_num_customer)*100
exist_customer_percent = (num_exist_customer / tot_num_customer)*100
# print text to show the distribution existing and churned customers in the dataset
print('The dataset consists of {:.2f}% which are Existing Customers and {:.2f}% which are customers who have churned / left the bank.'.format(exist_customer_percent, att_customer_percent))
# visualise this using a pie chart
Q1_df["Attrition_Flag"].value_counts().plot(kind="pie", y="Attrition_Flag", autopct="%1.1f%%", colors=["slateblue", "orchid"])
```

```python
# determine the distribution of gender in the number of exiting and churned customers
# for the churned customers
att_customer_M = CCCP_data[(CCCP_data.Attrition_Flag=="Attrited Customer") & (CCCP_data.Gender=="M")].Attrition_Flag.count()
churned_M = (att_customer_M / num_att_customer)*100
att_customer_F = CCCP_data[(CCCP_data.Attrition_Flag=="Attrited Customer") & (CCCP_data.Gender=="F")].Attrition_Flag.count()
churned_F = (att_customer_F / num_att_customer)*100
# for the existing customers
exist_customer_M = CCCP_data[(CCCP_data.Attrition_Flag=="Existing Customer") & (CCCP_data.Gender=="M")].Attrition_Flag.count()
exist_M = (exist_customer_M / num_exist_customer)*100
exist_customer_F = CCCP_data[(CCCP_data.Attrition_Flag=="Existing Customer") & (CCCP_data.Gender=="F")].Attrition_Flag.count()
exist_F = (exist_customer_F / num_exist_customer)*100
# print text to show the distritbution of gender in the existing and churned customers in the dataset
print('Within the Existing Customers {:.2f}% are Male and {:.2f}% Female'.format(exist_M, exist_F))
print('and within the Churned Customers {:.2f}% are Male and {:.2f}% Female'.format(churned_M, churned_F))

# to visulaise this dataset
customer_status = list(CCCP_data.Attrition_Flag.unique())
Male = [churned_M, exist_M]
Female = [churned_F, exist_F]

width = 0.40
plt.subplot(1, 2, 2)
index = np.arange(len(customer_status))
index_2 = [i + width for i in index]

plt.bar(index, Male, width=width, color = 'lightblue', label = "Male")
plt.bar(index_2, Female, width=width, color = 'pink', label = "Female")
plt.xticks(index, customer_status, color = 'black', rotation = 30)
plt.legend()

```
